```python
# Here we are creating a dataframe to append stories 
def get_stories():
    df['polarity'] = np.nan 
    df['subjectivity'] = np.nan
    df['Score'] = np.nan
    df['story'] = np.nan
    df["normalized"] = np.nan
    #SLCN_df["pos"] = np.nan
    #SLCN_df["neg"] = np.nan
    #SLCN_df["neu"] = np.nan
    for row, storyId in enumerate(df['storyId'].values):  #for each row in our df dataframe
        newsText = ek.get_news_story(storyId) #get the news story
        if newsText:
            soup = BeautifulSoup(newsText,"lxml") #create a BeautifulSoup object from our HTML news article
            sentA = TextBlob(soup.get_text()) #pass the text only article to TextBlob to anaylse
            story = html.document_fromstring(newsText).text_content() #after trial and error this provided string like story
            df['polarity'].iloc[row] = sentA.sentiment.polarity #write sentiment polarity back to df
            df['subjectivity'].iloc[row] = sentA.sentiment.subjectivity #write sentiment subjectivity score back to df
            df['story'].iloc[row] = story #Append Stories 
            #SLCN_df['story'].iloc[row] = ''.join(sentA)
            if sentA.sentiment.polarity >= 0.05: # attribute bucket to sentiment polartiy
                score = 'positive'
            elif  -.05 < sentA.sentiment.polarity < 0.05:
                score = 'neutral'
            else:
                score = 'negative'
            df['Score'].iloc[row] = score #write score back to df
            if sentA.sentiment.polarity  >= 0.05:  # Positive
                normalized = 1
            elif  -.05 < sentA.sentiment.polarity < 0.05:# Neutral
                normalized = -1
            else:
                normalized = 0  # Negative
            df["normalized"].iloc[row] = normalized
            df.head()
```

```python
lemmatizer = WordNetLemmatizer()
def tokenizer(row):
    """Tokenizes text.""" 
    #stop_words = sw
     #set(stopwords.words('english'))
    #sw_addon = {'to','a','capital, iq','in', 'the','per','to','and', 
              #  'on','sp','is','at','from','httptmsnrtrsjvdmxllme','fy','wa','ha','bordercolorinherit','storycontent',
              # 'to','a','capital, iq','in', 'the','per','to','and', 'on','sp','is','at','from','iq','v','co','mln','importantoutlinecolorinherit',
              # 'Copyright','mm','miss','q','bln','importantclick','link','video','iii','directory','thought','together','fts','praet','run','bit'}
    regex = re.compile("[^a-zA-Z ]")
    re_clean = regex.sub('', row)
    words = word_tokenize(re_clean)
    lem = [lemmatizer.lemmatize(word) for word in words]
    output = [word.lower() for word in lem if word.lower() not in sw]
    return output
```

```python
# Use the token_count function to generate the top 10 words from each coin
def token_count(tokens, N=1000):
    bigrams = ngrams(tokens,n=2)
    top_10 = dict(Counter(bigrams).most_common(N))
    return pd.DataFrame(list(top_10.items()), columns=['bigram', 'count'])
```

```python
def bigram_counter(tokens):
    bigram_count= dict(Counter(ngrams(tokens, n=2)))
    return pd.DataFrame(list(bigram_count.items()), columns=['bigram', 'count'])
```
